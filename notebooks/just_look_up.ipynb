```python
#clalcualte uniues of the columns 1 is all unique 0 is not unique and also the numbers need to be compared until the last digit
def unique_values(df):
    unique_counts = {}
    for column in df.columns:
        unique_counts[column] = df[column].nunique()
    return unique_counts
unique_counts = unique_values(df_last)
print(unique_counts)
#now the % of unique values
def unique_values_percent(df):
    unique_counts = {}
    for column in df.columns:
        unique_counts[column] = df[column].nunique() / len(df) * 100
    return unique_counts
unique_counts_percent = unique_values_percent(df_last)
print(unique_counts_percent)
```

```python
#print all similar columns in % below 5%
def similar_columns(df, threshold=0.05):
    similar_cols = {}
    for column in df.columns:
        if df[column].nunique() / len(df) < threshold:
            similar_cols[column] = df[column].nunique() / len(df) * 100
    return similar_cols
similar_cols = similar_columns(df_last)
print(similar_cols)
```
