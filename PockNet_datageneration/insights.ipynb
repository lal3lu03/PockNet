```python
# Strategies for handling NaN values
print("=== NaN Value Handling Strategies ===")
print("\n1. REMOVE rows with NaN values:")
print(f"   BU48: Would remove {len(nan_rows_bu48)} rows ({len(nan_rows_bu48)/len(df_2)*100:.2f}% of data)")
print(f"   Chen11: Would remove {len(nan_rows_chen11)} rows ({len(nan_rows_chen11)/len(df)*100:.2f}% of data)")

print("\n2. IMPUTE with zeros (conservative approach):")
print("   - Chemical features for UNK residues set to 0")
print("   - Assumes unknown residues have no chemical properties")

print("\n3. IMPUTE with mean/median values:")
print("   - Use mean/median of known chemical features")
print("   - May introduce bias")

print("\n4. SEPARATE handling:")
print("   - Train separate models for UNK vs known residues")
print("   - More complex but potentially more accurate")

print("\nRECOMMENDATION:")
print("Since NaN values are concentrated in UNK residues (unknown/hetero atoms),")
print("and chemical features are undefined for these, the most reasonable approach")
print("is to IMPUTE with zeros, treating UNK residues as chemically neutral.")
```

```python
import pandas as pd
# find all files according to the joind.ds with all files containing in this folder output_train/. the csv files have the name of the pdb file in the .ds file + _features.csv. 
#creat a new df with all this files combined to one df
# in the .ds file are all paths to the pdb files 
import os
def find_csv_files_in_output_train():
    output_train_path = "output_train"
    csv_files = []
    
    for root, dirs, files in os.walk(output_train_path):
        for file in files:
            if file.endswith("_features.csv"):
                csv_files.append(os.path.join(root, file))
    
    return csv_files

# only combine the files which are in the joind.ds file
''' the ds contains like this few first lines:

# benchmark 2011 dataset

PARAM.LIGANDS_SEPARATED_BY_TER=true

# liganded-pdb-file 

../p2rank-datasets/joined/astex/1g9v.pdb
../p2rank-datasets/joined/astex/1gkc.pdb
.
.
.
'''
def combine_csv_files_from_ds(ds_file_path):
    # Read the .ds file to get the list of PDB files
    with open(ds_file_path, 'r') as ds_file:
        pdb_files = [line.strip() for line in ds_file if line.strip() and line.startswith('..')]
    
    # Create a list to hold DataFrames
    dataframes = []
    
    # Iterate through the PDB files and read the corresponding CSV files
    for pdb_file in pdb_files:
        csv_file = f"output_train/{os.path.basename(pdb_file)}_features.csv"
        if os.path.exists(csv_file):
            df = pd.read_csv(csv_file)
            dataframes.append(df)
        else:
            print(f"Warning: {csv_file} does not exist.")
    
    # Combine all DataFrames into one
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    return combined_df

# ds file path /system/user/studentwork/hageneder/MSC/Practical_work/PockNet_datageneration/joind.ds
ds_file_path = "/system/user/studentwork/hageneder/MSC/Practical_work/PockNet_datageneration/joind.ds"
combined_df = combine_csv_files_from_ds(ds_file_path)
# display the combined dataframe
combined_df.head()
```

```python
# Save the combined DataFrame to a CSV file as finetune.csv
combined_df.to_csv("output_train/finetune.csv", index=False)
```
